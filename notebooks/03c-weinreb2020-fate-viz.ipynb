```python
base_dir = "../ext/klein_Figure5_data_and_code/"

timepoints = np.load(base_dir + 'timepoints.npy')
neu_mo_mask = np.load(base_dir + 'neutrophil_monocyte_trajectory_mask.npy')
clonal_fate_matrix = np.load(base_dir + 'clonal_fate_matrix.npy')

PBA_predictions = np.load(base_dir + 'PBA_predictions.npy')
FateID_predictions = np.load(base_dir + 'FateID_predictions.npy')
WOT_predictions = np.load(base_dir + 'WOT_predictions.npy')

early_cells = np.load(base_dir + 'early_cells.npy')
heldout_mask = np.load(base_dir + 'heldout_mask.npy')

smoothed_groundtruth = np.load(base_dir + 'smoothed_groundtruth_from_heldout.npy')

has_fate_mask = np.all([clonal_fate_matrix[:,5:7].sum(1) > 0, neu_mo_mask[timepoints==2]],axis=0)
neu_vs_mo_percent = clonal_fate_matrix[has_fate_mask,5] / clonal_fate_matrix[has_fate_mask,5:7].sum(1)

heldout_mask_for_early_cells = (~heldout_mask)[early_cells[neu_mo_mask[timepoints==2]]]

predictions = []
predictions.append(smoothed_groundtruth[np.all([early_cells[neu_mo_mask[timepoints==2]], ~heldout_mask],axis=0)])

predictions.append(PBA_predictions[early_cells[neu_mo_mask[timepoints==2]]][heldout_mask_for_early_cells])
predictions.append(WOT_predictions[early_cells[neu_mo_mask[timepoints==2]]][heldout_mask_for_early_cells])
predictions.append(FateID_predictions[early_cells[neu_mo_mask[timepoints==2]]][heldout_mask_for_early_cells])

clonal_data_no_heldout = neu_vs_mo_percent[np.all([early_cells[neu_mo_mask[timepoints==2]], ~heldout_mask],axis=0)[has_fate_mask[neu_mo_mask[timepoints==2]]]]

has_fate_mask_for_early_heldout = has_fate_mask[np.all([neu_mo_mask[timepoints==2],early_cells],axis=0)][heldout_mask_for_early_cells]
predictions_filtered = [pred[has_fate_mask_for_early_heldout] for pred in predictions]
```

# Examples of trajectories sampled from a model

```python
base_dir = "../experiments/weinreb-fate-train_batch-0.1/all_kegg-softplus_2_400-1e-06/"
config_path = os.path.join(base_dir, 'seed_2/config.pt'.format(1))
config = SimpleNamespace(**torch.load(config_path))

fate_pt = torch.load("../" + os.path.join(config.out_dir, 'fate.pt'))

scores = fate_pt['scores']
mask = fate_pt['mask']
names = fate_pt['names']

ensembled_scores = np.stack(scores[-1:], axis = 0).mean(axis = 0)
n = np.array(mask[-1:]).any(axis = 0).sum() 
n
```

```python
device = torch.device('cuda:0')
kwargs = {'num_workers': 1, 'pin_memory': True}
```

```python
model = this.AutoGenerator(config)
```

```python
checkpoint = torch.load("../" + os.path.join(config.out_dir, 'train.epoch_002500.pt'))
model.load_state_dict(checkpoint['model_state_dict'])
model.to(device)
```

```python
ns = 50
x_i = x[-11].expand(ns, -1).to(device)
xus_i = []

for _ in range(100): # step far beyond time steps
    z = torch.randn(x_i.shape[0], x_i.shape[1]) * config.train_sd
    z = z.to(device)
    x_i = model._step(x_i, dt = 0.1, z = z)
    
    x_i_ = x_i.detach().cpu().numpy()
    xu_i_ = um.transform(x_i_)
    xus_i.append(xu_i_)
    
xus = np.stack(xus_i)
c = np.arange(len(xus_i))    

fig, ax = plt.subplots(figsize = (6, 6))
for i in range(len(xu_)): 
    ax.scatter(xu_[i][:,0], xu_[i][:,1], s = 1, color = 'gray')    

for i in range(ns): 
    ax.scatter(xus[:,i,0], xus[:,i,1], c = c)
    ax.plot(xus[:,i,0], xus[:,i,1], '-', color = 'k', linewidth = 0.2)
```

# Visualize landscape

```python
base_dir = "../experiments/weinreb-fate-train_batch-0.1/all_kegg-softplus_2_400-1e-06/"
config_path = os.path.join(base_dir, 'seed_2/config.pt'.format(1))
config = SimpleNamespace(**torch.load(config_path))

fate_pt = torch.load("../" + os.path.join(config.out_dir, 'fate.pt'))

scores = fate_pt['scores']
mask = fate_pt['mask']
names = fate_pt['names']

ensembled_scores = np.stack(scores[-1:], axis = 0).mean(axis = 0)
n = np.array(mask[-1:]).any(axis = 0).sum() 
n
```

```python
device = torch.device('cuda:0')
kwargs = {'num_workers': 1, 'pin_memory': True}

model = this.AutoGenerator(config)
```

```python
checkpoint = torch.load("../" + os.path.join(config.out_dir, 'train.epoch_002500.pt'))
model.load_state_dict(checkpoint['model_state_dict'])
model.to(device)
```

Randomly sampling points

```python
fig, ax = plt.subplots(figsize = (6, 6))
for i in range(len(xu_)): 
    ax.scatter(xu_[i][:,0], xu_[i][:,1], s = 1, color = 'gray')  

for i in range(2): 

    xv = xp_[i][:100,:].to(device)
    xv = xv + model._drift(xv)
    xu_u = xu_[i][:100,:].numpy()
    xv_v = um.transform(xv.detach().cpu().numpy()) - xu_u
    xv_v = xv_v / (np.linalg.norm(xv_v, axis = 1)[:,np.newaxis])

    ax.quiver(xu_u[:,0], xu_u[:,1], xv_v[:,0], xv_v[:,1])
```

Sample uniformly on umap grid by chopping up umap grid

```python
xg = 5
yg = 20
xgrid = np.linspace(-10, 10, xg)
ygrid = np.linspace(-10, 10, yg)

fig, ax = plt.subplots(figsize = (6, 6))
for i in range(len(xu_)): 
    ax.scatter(xu_[i][:,0], xu_[i][:,1], s = 1, color = 'gray') 
    

for xi in range(xg-1): 
    for yi in range(yg-1): 
        
        xmin, xmax = xgrid[xi], xgrid[xi+1]
        ymin, ymax = ygrid[yi], ygrid[yi+1]
    
        in_x = (xu[:,0] > xmin) & (xu[:,0] < xmax)
        in_y = (xu[:,1] > ymin) & (xu[:,1] < ymax)
        in_box = in_x & in_y    
        
        if in_box.sum() > 3: 
            ix = np.random.choice(np.where(in_box)[0], max(3, int(in_box.sum() * 0.002)))

            xv = torch.from_numpy(xp[ix,:]).float().to(device)
            xv = xv + model._drift(xv)
            xv = um.transform(xv.detach().cpu().numpy()) - xu[ix,:]
            xv = xv / np.linalg.norm(xv, axis = 1)[:,np.newaxis]

            ax.quiver(xu[ix,0], xu[ix,1], xv[:,0], xv[:,1], scale = 0.9, scale_units = 'xy', 
                width = .004)
```
