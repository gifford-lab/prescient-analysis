# Load data for visualization

```python
expr = sh.load_npz("/data/gl/g2/sachit/data/klein/processed/counts_matrix_in_vitro.npz")
genes = sh.load_genes("/data/gl/g2/sachit/data/klein/processed/gene_names_in_vitro.txt")
highly_variable_genes = sh.filter_genes(expr)
cell_cycle_genes = [['Ube2c',
                    'Hmgb2',
                    'Hmgn2',
                    'Tuba1b', 
                    'Ccnb1', 
                    'Tubb5', 
                    'Top2a', 
                    'Tubb4b']]
non_cell_cycle_idx = sh.remove_corr_genes(expr, gene_list=genes, exclude_corr_genes_list=cell_cycle_genes, test_gene_idx=highly_variable_genes, min_corr=0.1)
non_cell_cycle_idx = non_cell_cycle_idx[0:2447]
# perform filtering
expr = expr[:, non_cell_cycle_idx]

use_genes = [genes[i] for i in non_cell_cycle_idx]
```

```python
meta = pd.read_csv("/data/gl/g2/sachit/data/klein/spring/in_vitro/metadata.csv")
meta = meta.drop(meta.columns[0], axis = 1)
meta.head()
```

```python
x = expr.todense()
```

```python
y = meta['Time point'].values.astype(int)
```

```python
ix_tr = meta['Time point'] != 4
ix_te = meta['Time point'] == 4

x_tr = x[ix_tr]
x_te = x[ix_te]
y_tr = y[ix_tr]
y_te = y[ix_te]

meta_tr = meta.loc[ix_tr]
meta_te = meta.loc[ix_te]
```

```python
scaler = sklearn.preprocessing.StandardScaler()
pca = sklearn.decomposition.PCA(n_components = 50)
um = umap.UMAP(n_components = 2, metric = 'euclidean', n_neighbors = 30)
```

```python
x_tr = scaler.fit_transform(x_tr)
xp_tr = pca.fit_transform(x_tr)
xu_tr = um.fit_transform(xp_tr) 

x_te = scaler.transform(x_te)
xp_te = pca.transform(x_te)
xu_te = um.transform(xp_te)
```

```python
x_l = [x_tr[y_tr == 2,], x_te[y_te == 4,], x_tr[y_tr == 6,]]
xp_l = [xp_tr[y_tr == 2,], xp_te[y_te == 4,], xp_tr[y_tr == 6,]]
xu_l = [xu_tr[y_tr == 2,], xu_te[y_te == 4,], xu_tr[y_tr == 6,]]

x_l = [torch.from_numpy(a).float() for a in x_l]
xp_l = [torch.from_numpy(a).float() for a in xp_l]
xu_l = [torch.from_numpy(a).float() for a in xu_l]
```

```python
xp = torch.cat(xp_l).cpu().numpy()
xu = torch.cat(xu_l).cpu().numpy()
xp.shape, xu.shape
```

# Load model (d26 weights)

```python
config_path = ("../experiments/weinreb-interpolate-train_batch-0.1/d26-softplus_2_400-0.1_0.1_1e-06-0.1_0.1_0.005/"
            "seed_4/config.pt")
config = SimpleNamespace(**torch.load(config_path))
config
```

```python
device = torch.device('cuda:0')
kwargs = {'num_workers': 1, 'pin_memory': True}
```

```python
log_path = config_path.replace('config.pt', 'interpolate.log')

log = pd.read_csv(log_path, sep = '\t')
log = log.loc[log['epoch'] != 'best']
log['epoch'] = log['epoch'].apply(lambda e: int(e.split('_')[1]))

log = log.set_index(['epoch', 'eval'])['loss'].unstack()
best = log.sort_values('train').head(n = 1)
best
```

```python
x, y, w = weinreb.load_data(config, base_dir = "..")
```

```python
model = train.AutoGenerator(config)
```

```python
train_pt = "../" + config.train_pt.format('epoch_002100')
checkpoint = torch.load(train_pt)
model.load_state_dict(checkpoint['model_state_dict'])
model.to(device)
print(model)
```

## day 4

```python
evaluate_n = 10000

t_prev = config.start_t
t_cur = 1 # day 4

y_prev = int(y[t_prev])
y_cur = int(y[t_cur])

time_elapsed = y_cur - y_prev
num_steps = int(np.round(time_elapsed / config.train_dt))

dat_prev = x[t_prev].to(device)
dat_cur = x[t_cur].to(device)
w_prev = train.get_weight(w[(y_prev, y_cur)], time_elapsed).cpu().numpy()

x_s = []
x_i_ = train.weighted_samp(dat_prev, evaluate_n, w_prev)

for i in range(int(evaluate_n / config.ns)):

    x_i = x_i_[i*config.ns:(i+1)*config.ns,]

    for _ in range(num_steps):
        z = torch.randn(x_i.shape[0], x_i.shape[1]) * config.train_sd
        z = z.to(device)
        x_i = model._step(x_i, dt = config.train_dt, z = z)

    x_s.append(x_i.detach())

x_s = torch.cat(x_s).cpu().numpy()
xu_s = um.transform(x_s)
```

```python
fig, ax = plt.subplots(figsize = (6, 6))
ax.scatter(xu_tr[:,0], xu_tr[:,1], s = 1, color = 'gray')
ax.scatter(xu_te[:,0], xu_te[:,1], s = 1, color = 'gray')

ax.scatter(xu_s[:,0], xu_s[:,1], color = 'red', s = 1)
sns.kdeplot(xu_s[:,0], xu_s[:,1], color = 'k')

xlim = ax.get_xlim()
ylim = ax.get_ylim()

ax.set_xlim(xlim)
ax.set_ylim(ylim)
```

Visualize potential + landscape

```python
xg = 10
yg = 20
xgrid = np.linspace(-10, 10, xg)
ygrid = np.linspace(-10, 10, yg)

fig, ax = plt.subplots(figsize = (6,6))

ax.scatter(xu[:,0], xu[:,1], s = 1, color = 'gray')

for xi in range(xg-1): 
    for yi in range(yg-1): 
        
        xmin, xmax = xgrid[xi], xgrid[xi+1]
        ymin, ymax = ygrid[yi], ygrid[yi+1]
    
        in_x = (xu[:,0] > xmin) & (xu[:,0] < xmax)
        in_y = (xu[:,1] > ymin) & (xu[:,1] < ymax)
        in_box = in_x & in_y    
        
        if in_box.sum() > 3: 
            ix = np.random.choice(np.where(in_box)[0], max(3, int(in_box.sum() * 0.002)))

            xv = torch.from_numpy(xp[ix,:]).float().to(device)
            xv = xv + model._drift(xv)
            xv = um.transform(xv.detach().cpu().numpy()) - xu[ix,:]
            xv = xv / np.linalg.norm(xv, axis = 1)[:,np.newaxis]

            ax.quiver(xu[ix,0], xu[ix,1], xv[:,0], xv[:,1], scale = 0.9, scale_units = 'xy', 
                width = .004)
```

```python
fig, ax = plt.subplots(figsize = (7, 6))
xc = []

c = model._pot(torch.from_numpy(xp).float().to(device)).detach().cpu().numpy()[:,0] * -1
ci = np.argsort(c)

sax = ax.scatter(xu[ci,0], xu[ci,1], s = 1, c = c[ci], vmin = 0, vmax = 300)

plt.colorbar(sax, shrink = 0.9)    
```
