```python
x = pd.read_csv("../data/Veres2019/Stage_5.Seurat.csv", index_col = 0)
x.head()
```

```python
scaler = sklearn.preprocessing.StandardScaler()
xs = pd.DataFrame(scaler.fit_transform(x), index = x.index, columns = x.columns)
```

```python
x1 = xs['TOP2A']
rs = {}
for g in xs.columns: 
    
    x2 = xs[g]
    r, pval = scipy.stats.pearsonr(x1, x2)

    rs[g] = r
rs = pd.Series(rs)
(rs > 0.15).value_counts()
```

Use only genes not correlated with TOP2A as input into PCA for visualization and model fitting

```python
pca = sklearn.decomposition.PCA(n_components = 30)

xs_f = xs.loc[:,rs < 0.15]

xp = pd.DataFrame(pca.fit_transform(xs_f), index = xs.index)
xp.head()
```

Compute the umap and plot

```python
um = umap.UMAP(n_components = 2, metric = 'euclidean', n_neighbors = 30)
xu = um.fit_transform(xp)
```

# Load KEGG annotations

```python
gst = pd.read_csv("./depr/grace/growth/hs_birth_msigdb_kegg.csv", index_col = 0)
birth_gst = [g for g in gst['gene_symbol'].unique() if g in x.columns]

gst = pd.read_csv("./depr/grace/growth/hs_death_msigdb_kegg.csv", index_col = 0)
death_gst = [g for g in gst['gene_symbol'].unique() if g in x.columns]

birth_gst = [g for g in birth_gst if g not in death_gst]
death_gst = [g for g in death_gst if g not in birth_gst]

len(birth_gst), len(death_gst)
```

For smoothing we will use all genes

```python
pca = sklearn.decomposition.PCA(n_components = 30)
xp_ = pca.fit_transform(xs)
```

```python
ay = annoy.AnnoyIndex(xp_.shape[1], 'euclidean') 
for i in range(xp_.shape[0]):
    ay.add_item(i, xp_[i])
ay.build(10)
```

```python
n_neighbors = 20
beta = 0.1

# smooth growth score

prev_score = xs[birth_gst].mean(axis = 1).values
cur_score = np.zeros(prev_score.shape)

for _ in range(5): 
    for i in range(len(prev_score)): 
        xn = prev_score[ay.get_nns_by_item(i, 20)]
        cur_score[i] = (beta * xn[0]) + ((1 - beta) * xn[1:].mean(axis = 0))
    prev_score = cur_score
    
birth_score = xs[birth_gst].mean(axis = 1).values
birth_smoothed_score = cur_score

# smooth death score

prev_score = xs[death_gst].mean(axis = 1)
cur_score = np.zeros(prev_score.shape)

for _ in range(5): 
    for i in range(len(prev_score)): 
        xn = prev_score[ay.get_nns_by_item(i, 20)]
        cur_score[i] = (beta * xn[0]) + ((1 - beta) * xn[1:].mean(axis = 0))
    prev_score = cur_score
    
death_score = xs[death_gst].mean(axis = 1).values  
death_smoothed_score = cur_score
```

```python
L0 = 0.3
L = 1.1
k = 0.001

kb = np.log(k) / np.min(birth_score)
kd = np.log(k) / np.min(death_score)

b = birth_smoothed_score
d = death_smoothed_score

b = L0 + L / (1 + np.exp(-kb * b))
d = L0 + L / (1 + np.exp(-kd * d))
g = b - d

np.exp(g.min()), np.exp(g.max())
```

```python
fig, ax = plt.subplots(figsize = (6,6))

c = np.exp(g)
ci = np.argsort(c)
sax = ax.scatter(xu[ci,0], xu[ci,1], s = 1, c = c[ci])

plt.colorbar(sax, shrink = 0.9)    
```
