From the paper

> Dimensionality reduction and clus-tering for each dataset was performed by broadly following a modified version of a previously published approach. Using the unnormalized counts, highly varia-ble genes were identified as previously described, by finding outliers with high coefficients of variations as a function of mean expression. Then, within each dataset, depth-normalized counts values were further z-normalized per gene, to yield z-normalized values. The z-normalized values of variable genes per dataset were used as input for principal component analysis. When computing principal components for the stage-5 datasets, we identified genes correlated with cell-cycle marker TOP2A (Pearson correlation greater 0.15), and excluded them. Clustering was carried out using Leiden community detection, a recently published improve-ment on Louvain community detection.

Load the unscaled but normalized data for the top 2500 highly variable genes

```python
x = pd.read_csv("../data/Veres2019/Stage_5.Seurat.csv", index_col = 0)
x.head()
```

Then, scale the data

```python
scaler = sklearn.preprocessing.StandardScaler()
xs = pd.DataFrame(scaler.fit_transform(x), index = x.index, columns = x.columns)
```

Identify genes correlated with TOP2A

```python
x1 = xs['TOP2A']
rs = {}
for g in xs.columns: 
    
    x2 = xs[g]
    r, pval = scipy.stats.pearsonr(x1, x2)

    rs[g] = r
rs = pd.Series(rs)
(rs > 0.15).value_counts()
```

Use only genes that are not correlated with TOP2A as input into PCA

```python
pca = sklearn.decomposition.PCA(n_components = 30)

xs_f = xs.loc[:,rs < 0.15]

xp = pd.DataFrame(pca.fit_transform(xs_f), index = xs.index)
xp.head()
```

Compute the umap and plot

```python
um = umap.UMAP(n_components = 2, metric = 'euclidean', n_neighbors = 30)
xu = um.fit_transform(xp)
```

Visualize time, it actually looks like the system approaches steady state at day 3

```python
evaluate_n = 10000

t_prev = config.start_t
t_cur = 1 

y_prev = int(y[t_prev])
y_cur = int(y[t_cur])

time_elapsed = y_cur - y_prev
num_steps = int(np.round(time_elapsed / config.train_dt))

dat_prev = x[t_prev].to(device)
dat_cur = x[t_cur].to(device)
w_prev = train.get_weight(w[(y_prev, y_cur)], time_elapsed).cpu().numpy()

x_s = []
x_i_ = train.weighted_samp(dat_prev, evaluate_n, w_prev)

for i in range(int(evaluate_n / config.ns)):

    x_i = x_i_[i*config.ns:(i+1)*config.ns,]

    for _ in range(num_steps):
        z = torch.randn(x_i.shape[0], x_i.shape[1]) * config.train_sd
        z = z.to(device)
        x_i = model._step(x_i, dt = config.train_dt, z = z)

    x_s.append(x_i.detach())

x_s = torch.cat(x_s).cpu().numpy()
xu_s = um.transform(x_s)
```

Visualize landscape by randomly sampling points

```python
xg = 20
yg = 20
xgrid = np.linspace(-10, 10, xg)
ygrid = np.linspace(-5, 5, yg)

fig, ax = plt.subplots(figsize = (6,6))

ax.scatter(xu[:,0], xu[:,1], s = 1, color = 'gray')

for xi in range(xg-1): 
    for yi in range(yg-1): 
        
        xmin, xmax = xgrid[xi], xgrid[xi+1]
        ymin, ymax = ygrid[yi], ygrid[yi+1]
    
        in_x = (xu[:,0] > xmin) & (xu[:,0] < xmax)
        in_y = (xu[:,1] > ymin) & (xu[:,1] < ymax)
        in_box = in_x & in_y    
        
        if in_box.sum() > 3: 
            ix = np.random.choice(np.where(in_box)[0], max(3, int(in_box.sum() * 0.002)))

            xv = torch.from_numpy(xp.values[ix,:]).float().to(device)
            xv = xv + model._drift(xv)
            xv = um.transform(xv.detach().cpu().numpy()) - xu[ix,:]
            xv = xv / np.linalg.norm(xv, axis = 1)[:,np.newaxis]

            ax.quiver(xu[ix,0], xu[ix,1], xv[:,0], xv[:,1], scale = 1.5, scale_units = 'xy', 
                width = .004)
```

```python
fig, ax = plt.subplots(figsize = (7, 6))
xc = []

c = model._pot(torch.from_numpy(xp.values).float().to(device)).detach().cpu().numpy()[:,0] * -1
ci = np.argsort(c)

sax = ax.scatter(xu[ci,0], xu[ci,1], s = 1, c = c[ci], vmax = 1000)

plt.colorbar(sax, shrink = 0.9)    
```

# Cell type classifier

Fit a cell type classifier to predict sc-beta, sc-alpha, sc-ec vs everything else

```python
yc = metadata['Assigned_cluster']
yc = yc.mask(~((yc == 'sc_alpha') | (yc == 'sc_beta') | (yc == 'sc_ec')), 'Other')
yc.value_counts()
```

```python
xtr, xte, ytr, yte = sklearn.model_selection.train_test_split(xp.values, yc, stratify = yc)
```

```python
n_trees = 10
n_neighbors = 20

t = annoy.AnnoyIndex(xtr.shape[1], 'euclidean')
for i in range(xtr.shape[0]): 
    t.add_item(i, xtr[i])
t.build(n_trees)
```

```python
yp = []
for i in range(xte.shape[0]): 
    nn = ytr.iloc[t.get_nns_by_vector(xte[i], n_neighbors)]
    nn = Counter(nn.values).most_common(2)
    label, num = nn[0]
    yp.append(label)
```

```python
print(sklearn.metrics.classification_report(yte, yp))
```

```python
sklearn.metrics.f1_score(yte, yp, average = 'macro')
```
