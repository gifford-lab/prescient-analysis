Training curves for 2_400

```python
fig, axs = plt.subplots(5, 2, figsize = (10, 10))

paths = glob.glob("../experiments/weinreb-interpolate-train_batch-0.1/"
          "/d26-softplus_2_400-0.1_0.1_1e-06-0.1_0.1_0.005/*/interpolate.log")

for ai, path in enumerate(sorted(paths)): 
    
    log = pd.read_csv(path, sep = '\t')
    log = log.loc[log['epoch'] != 'best']
    log['epoch'] = log['epoch'].apply(lambda e: int(e.split('_')[1]))

    ax = axs[ai, 0]
    log_ = log.loc[log['eval'] == 'train']
    ax.plot(log_['epoch'], log_['loss'], 'b-')

    ax2 = ax.twinx()
    log_ = log.loc[log['eval'] == 'test']
    ax2.plot(log_['epoch'], log_['loss'], 'r-')
    
    log = log.set_index(['epoch', 'eval'])['loss'].unstack()
    best = log.sort_values('train').head(n = 1)
    best_train = best['train'].item()
    best_test = best['test'].item()
    
    ax.axvline(best.index.item(), color = 'k', linestyle = 'dashed')
    ax.set_title('Train: {:.3f}, Test: {:.3f}'.format(best_train, best_test))
    
    ax2.spines["left"].set_color('b')
    ax2.spines["right"].set_color('r')     
    
paths = glob.glob("../experiments/weinreb-interpolate-train_batch-0.1/"
          "/d26-softplus_2_400-0.1_0.1_1e-06-0.1_0.1_0.01/*/interpolate.log")    
    
for ai, path in enumerate(sorted(paths)): 
    
    log = pd.read_csv(path, sep = '\t')
    log = log.loc[log['epoch'] != 'best']
    log['epoch'] = log['epoch'].apply(lambda e: int(e.split('_')[1]))

    ax = axs[ai, 1]
    log_ = log.loc[log['eval'] == 'train']
    ax.plot(log_['epoch'], log_['loss'], 'b-')

    ax2 = ax.twinx()
    log_ = log.loc[log['eval'] == 'test']
    ax2.plot(log_['epoch'], log_['loss'], 'r-')
    
    log = log.set_index(['epoch', 'eval'])['loss'].unstack()
    best = log.sort_values('train').head(n = 1)
    best_train = best['train'].item()
    best_test = best['test'].item()
    
    ax.axvline(best.index.item(), color = 'k', linestyle = 'dashed')
    ax.set_title('Train: {:.3f}, Test: {:.3f}'.format(best_train, best_test))    
    
    ax2.spines["left"].set_color('b')
    ax2.spines["right"].set_color('r')      
    
plt.tight_layout()
```
